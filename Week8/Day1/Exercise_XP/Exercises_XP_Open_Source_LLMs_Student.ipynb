**Deliverables:** paragraph + healthcare answer.

```python
# Définitions et caractéristiques clés des niveaux d'ouverture
open_source_levels = {
    "Fully Open": {
        "definition": "Modèles dont le code source, les données d'entraînement, les poids et les protocoles sont publics.",
        "what_is_open": ["Poids du modèle", "Code d'entraînement", "Données d'entraînement (Data)"],
        "what_you_can_do": ["Audit complet des données", "Reproduction totale", "Entraînement à partir de zéro"],
        "what_you_cannot_do": ["Utilisation sans attribution (selon licence)", "Violer les restrictions éthiques spécifiques"],
    },
    "Weights Released": {
        "definition": "Modèles dont seuls les poids finaux et l'architecture sont partagés, sans les données d'entraînement.",
        "what_is_open": ["Poids du modèle", "Architecture du réseau"],
        "what_you_can_do": ["Fine-tuning local", "Déploiement privé", "Inférence sur matériel propre"],
        "what_you_cannot_do": ["Savoir exactement sur quoi le modèle a été entraîné", "Vérifier les biais des données sources"],
    },
    "Architecture Only": {
        "definition": "Seule la structure du modèle est publiée. Les poids doivent être entraînés par l'utilisateur.",
        "what_is_open": ["Code de l'architecture", "Hyperparamètres"],
        "what_is_open": ["Architecture du réseau"],
        "what_you_can_do": ["Coder sa propre version", "Recherche académique sur la structure"],
        "what_you_cannot_do": ["Utiliser le modèle immédiatement", "Économiser les coûts de calcul d'entraînement"],
    },
}
open_source_levels
```

```python
# Remplissage de la table de comparaison
comparison_table = """| Openness level | What's open? | Impact on retraining/modifying |
| --- | --- | --- |
| Fully Open | Data, Code, Weights | Maximum: Allows full domain adaptation and safety audit. |
| Weights Released | Weights, Arch | High: Great for fine-tuning, but data lineage is unknown. |
| Architecture Only | Design/Code | Low: Requires massive compute to make it useful. """
print(comparison_table)
```

```python
# Paragraphe comparatif de 3 à 5 phrases
comparative_paragraph = """Le choix entre un modèle 'Fully Open' et un modèle aux poids simplement libérés ('Weights Released') dépend principalement du besoin de transparence et de contrôle sur les données. Alors qu'un modèle aux poids libérés comme Llama permet un déploiement rapide et efficace, seul un modèle totalement ouvert permet un audit de sécurité complet et une reproduction scientifique. En revanche, les publications limitées à l'architecture ('Architecture Only') servent principalement à la recherche théorique faute de poids pré-entraînés exploitables immédiatement. Ainsi, l'équilibre se trouve souvent entre la commodité des poids déjà calculés et la souveraineté totale des données sources."""
print(comparative_paragraph)
```
