Exercice 1 : Comprendre la visualisation des données

- Pourquoi la visualisation des données est importante dans l'analyse des données :
    La visualisation est importante parce qu’elle permet de mieux comprendre les données, de détecter des tendances et anomalies, de communiquer efficacement les résultats et d’aider à prendre de bonnes décisions.
    Sans visualisation, l’analyse des données serait beaucoup plus difficile et moins efficace.

- Décrire l'utilité d'un graphique linéaire en visualisation de données :
    Le graphique linéaire est très utile en visualisation de données car il permet de suivre les évolutions, comprendre les tendances et analyser les changements au fil du temps de manière claire et rapide.

Exercice 2 : Création d’un graphique linéaire pour la variation de température

```python
import matplotlib.pyplot as plt

jour = ["Lun", "Mar", "Mer", "Jeu", "Ven", "Sam", "Dim"]
temperature = [72, 74, 76, 80, 82, 78, 75]

plt.plot(jour, temperature)
plt.xlabel("Jour")
plt.ylabel("Température (°F)")
plt.title("Graphique linéaire pour la variation de température")
```

Exercice 3 : Visualisation des ventes mensuelles à l’aide d’un graphique à barres

```python

mois = ["Jan", "Fév", "Mar", "Avr", "Mai"]
Vente = [5000, 5500, 6200, 7000, 7500]

plt.bar(mois, Vente)
plt.xlabel("Mois")
plt.ylabel("Montant des ventes ($)")
plt.title("Visualisation des ventes mensuelles à l’aide d’un graphique à barres")
```

Exercice 5 : Comparaison des niveaux d’anxiété selon le sexe

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r"E:\FORMATION_Tech_Talent_Accelerator_de_initiative_FORPRODE\DI_Bootcamp\Week2\Day4\Exercise_XP\Student Mental health\Student Mental health.csv")
df = pd.DataFrame(data)

# Convertir "Yes" → 1 et "No" → 0
df["Anxiety_num"] = df["Do you have Anxiety?"].map({"Yes": 1, "No": 0})

sns.barplot( data= df, x="Choose your gender", y="Anxiety_num", palette= "Set2")
plt.title("Comparing Anxiety Levels Across Different Genders")
plt.xlabel("Genre")
plt.ylabel("Proportion avec anxiété")
plt.show()
```

Exercice 6 : Explorer la relation entre l’âge et les crises de panique

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement des données
data = pd.read_csv(r"E:\FORMATION_Tech_Talent_Accelerator_de_initiative_FORPRODE\DI_Bootcamp\Week2\Day4\Exercise_XP\Student Mental health\Student Mental health.csv")
df = pd.DataFrame(data)

# Convertir "Yes" → 1 et "No" → 0
df["Panic_Attack_num"] = df["Do you have Panic attack?"].map({"Yes": 1, "No": 0})

# Créer le nuage de points
sns.scatterplot(
    data=df,
    x="Age",
    y="Panic_Attack_num",
    hue="Panic_Attack_num",   # couleur selon 0 ou 1
    palette="coolwarm",
    s=100,                    # taille des points
    alpha=0.7                 # transparence
)

plt.title("Relation entre l'âge et les crises de panique")
plt.xlabel("Âge des étudiants")
plt.ylabel("Crise de panique (0 = Non, 1 = Oui)")
plt.show()
```
