Exercice 2 : Statistiques descriptives

```python
from scipy import stats

data = [12, 15, 13, 12, 18, 20, 22, 21]

# calcul des statistiques de base sur notre jeu de données
moyenne = stats.tmean(data)
mediane = stats.describe(data).mean
variance = stats.tvar(data)
ecart_type = stats.tstd(data)

print("Moyenne :", round(moyenne, 2))
print("Médiane :", mediane)
print("Variance :", round(variance, 2))
print("Ecart-type :", round(ecart_type, 2))
```

Exercice 3 : Distribution normale

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# on génère des valeurs entre 20 et 80 pour couvrir la distribution
x = np.linspace(20, 80, 100)

# on crée la courbe normale avec moyenne=50 et écart-type=10
y = norm.pdf(x, 50, 10)

# tracé du graphique
plt.plot(x, y)
plt.title("Distribution normale (moyenne=50, écart-type=10)")
plt.xlabel("Valeurs")
plt.ylabel("Probabilité")
plt.show()
```

Exercice 4 : Test T

```python
from scipy import stats
import numpy as np

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

# le test t compare les moyennes de deux groupes
# il nous dit si la différence entre les deux est significative ou due au hasard
t_stat, p_value = stats.ttest_ind(data1, data2)

print(f"Statistique T : {t_stat:.4f}")
print(f"Valeur P : {p_value:.4f}")

# si p < 0.05, la différence est significative
if p_value < 0.05:
    print("La différence entre les deux groupes est significative")
else:
    print("La différence entre les deux groupes n'est pas significative")
```

Exercice 5 : Régression linéaire

```python
from scipy import stats

house_sizes  = [50, 70, 80, 100, 120]
house_prices = [150000, 200000, 210000, 250000, 280000]

# la régression linéaire trouve la droite qui passe au mieux entre tous les points
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

print(f"Pente : {slope:.2f}")
print(f"Ordonnée à l'origine : {intercept:.2f}")
print(f"R² : {r_value**2:.4f}")

# on prédit le prix d'une maison de 90m²
taille = 90
prix_predit = slope * taille + intercept
print(f"\nPrix prédit pour 90m² : {prix_predit:.2f}")

# interprétation de la pente
print(f"\nInterprétation : chaque m² supplémentaire augmente le prix de {slope:.2f} unités")
```

Exercice 6 : ANOVA

```python
from scipy import stats

fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# l'ANOVA compare les moyennes de plusieurs groupes en même temps
# ici on teste si les 3 engrais ont des effets différents sur la croissance
f_stat, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print(f"Valeur F : {f_stat:.4f}")
print(f"Valeur P : {p_value:.4f}")

if p_value < 0.05:
    print("Les engrais ont des effets significativement différents")
else:
    print("Les engrais n'ont pas d'effets significativement différents")

print("\nSi P > 0.05 : on ne peut pas conclure que les engrais sont différents")
print("Cela signifierait que les différences observées sont dues au hasard")
```

Exercice 7 : Distribution binomiale (facultatif)

```python
from scipy.stats import binom

# on calcule la probabilité d'obtenir exactement 5 faces en 10 lancers
# n=10 lancers, p=0.5 probabilité d'avoir face
n = 10
p = 0.5
k = 5

probabilite = binom.pmf(k, n, p)
print(f"Probabilité d'obtenir exactement {k} faces en {n} lancers : {probabilite:.4f}")
print(f"Soit environ {probabilite*100:.1f}%")
```

Exercice 8 : Corrélation (facultatif)

```python
import pandas as pd
from scipy import stats

data = pd.DataFrame({
    'age':    [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

# Pearson mesure la corrélation linéaire entre deux variables
pearson_corr, pearson_p = stats.pearsonr(data['age'], data['income'])

# Spearman mesure la corrélation basée sur le rang, plus robuste aux valeurs extrêmes
spearman_corr, spearman_p = stats.spearmanr(data['age'], data['income'])

print(f"Corrélation de Pearson  : {pearson_corr:.4f} (p={pearson_p:.4f})")
print(f"Corrélation de Spearman : {spearman_corr:.4f} (p={spearman_p:.4f})")

# une corrélation proche de 1 signifie que les deux variables augmentent ensemble
if pearson_corr > 0.8:
    print("Forte corrélation positive : quand l'âge augmente, le salaire augmente aussi")
```
