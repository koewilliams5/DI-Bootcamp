What is data analysis?
- Data analysis involves transforming raw information into useful information to better understand a situation.

Why is data analysis important in modern contexts? 
- Data analysis is important because it helps transform raw information into useful knowledge for smarter decisions and better results.

List and describe three areas where data analysis is applied today:
answer: 
- Les entreprises (Business):
    Les entreprises utilisent l’analyse de données pour comprendre les besoins des clients, améliorer les ventes et augmenter les profits.

- L’éducation
    Les écoles et universités utilisent l’analyse de données pour suivre les performances des étudiants et améliorer l’apprentissage.

- Type d’analyse : Analyse temporelle

```python
##Exercice 9 : Exporter un dataframe au format Excel et au format JSON.

import pandas as pd

data = {
    "Nom": ["Alice", "Bob", "Charlie"],
    "Age": [22, 25, 30]
}

dc =pd.DataFrame(data)

## Export Excel
dc.to_excel("fichier.xlsx", index=False)

## Export JSON
dc.to_json("fichier.json", orient="records")

print("Exportation réussie")

```
