Défi quotidien : Gestion et analyse des données en Python

```python
import pandas as pd
import numpy as np
import requests
import zipfile
import io
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

# chargement du dataset
url = "https://github.com/devtlv/Datasets-DA-Bootcamp-2-/raw/refs/heads/main/Week%204%20-%20Data%20Understanding/W4D4%20-%20Data%20Preprocessing%20&%20T/Data%20Science%20Job%20Salary%20dataset.zip"

response = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(response.content))
df = pd.read_csv(z.open(z.namelist()[0]))

display(df.head())


# 1. NORMALISATION MIN-MAX sur la colonne salary
scaler = MinMaxScaler()
df['salary_normalized'] = scaler.fit_transform(df[['salary']])

print("Normalisation Min-Max sur salary :")
print(f"Min avant : {df['salary'].min()}")
print(f"Max avant : {df['salary'].max()}")
print(f"Min après : {df['salary_normalized'].min():.2f}")
print(f"Max après : {df['salary_normalized'].max():.2f}")


# 2. RÉDUCTION DE DIMENSIONNALITÉ avec PCA
colonnes_numeriques = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(f"\nColonnes numériques utilisées pour PCA : {colonnes_numeriques}")

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df[colonnes_numeriques])
df_pca = pd.DataFrame(df_pca, columns=['Composante_1', 'Composante_2'])

print(f"Variance expliquée : {pca.explained_variance_ratio_.sum()*100:.1f}%")
display(df_pca.head())


# 3. AGRÉGATION par niveau d'expérience
agregation = df.groupby('experience_level')['salary'].agg(
    Salaire_Moyen='mean',
    Salaire_Median='median'
).round(2)

print("\nSalaire moyen et médian par niveau d'expérience :")
display(agregation)
```
