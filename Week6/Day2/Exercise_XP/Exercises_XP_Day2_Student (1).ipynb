## Exercise 1 : Traditional vs. Modern NLP

### 1) Tableau comparatif

| Aspect | Traditional NLP | Modern NLP |
|---|---|---|
| Feature Engineering | Manuelle : TF-IDF, n-grams, POS tags définis par des experts | Automatique : le modèle apprend ses propres features depuis les données brutes |

| Disadvantages | Rigide, feature engineering coûteux, ne capture pas le contexte | Coûteux en ressources, boîte noire, données d'entraînement massives |

### 2) Impact sur la scalabilité et l'efficacité

Le passage au NLP moderne a profondément transformé la scalabilité
des applications. Avant, chaque nouvelle tâche nécessitait de
repartir de zéro : collecter des données labellisées, concevoir
manuellement des features, entraîner un nouveau modèle. Aujourd'hui,
un seul modèle pré-entraîné comme BERT peut être adapté à des
dizaines de tâches différentes en quelques heures.

Les architectures Transformer permettent un traitement parallèle
des séquences, contrairement aux RNN séquentiels — ce qui accélère
massivement l'entraînement sur GPU. Le transfer learning réduit
drastiquement le besoin en données labellisées, rendant les
applications NLP accessibles même pour des domaines spécialisés
comme le droit ou la médecine en Côte d'Ivoire, où les données
annotées sont rares.

## Exercise 2 : LLM Architecture and Application Scenarios

### BERT
**Architecture** : Encoder-only, bidirectionnel. BERT lit la phrase
dans les deux sens simultanément (gauche→droite ET droite→gauche).
Il utilise le Masked Language Modeling (MLM) : certains mots sont
masqués et le modèle doit les deviner en utilisant tout le contexte.

**Application** : Analyse de sentiment sur des avis clients
(positif / négatif / neutre).

**Pourquoi** : La bidirectionnalité permet à BERT de comprendre
le contexte complet d'un mot. Pour "Ce produit n'est PAS mauvais",
il comprend que "PAS mauvais" = positif grâce au contexte des deux côtés.

---

### GPT
**Architecture** : Decoder-only, unidirectionnel (gauche→droite).
GPT utilise le Causal Language Modeling (CLM) : il prédit le prochain
mot en ne regardant que les mots précédents. C'est un modèle autorégressif.

**Application** : Chatbot conversationnel et génération de code.

**Pourquoi** : Générer du texte mot par mot de gauche à droite est
exactement ce que fait GPT naturellement. C'est parfait pour les
réponses de chatbot où on produit du texte séquentiellement.

---

### T5
**Architecture** : Encoder-Decoder. L'encoder lit et comprend le texte
source, le decoder génère le texte cible. Tout est formulé comme
une tâche "texte → texte" : traduction, résumé, Q&A, tous utilisent
le même format.

**Application** : Traduction automatique de documents techniques
(anglais → français).

**Pourquoi** : La traduction nécessite de comprendre complètement
la phrase source (encoder) avant de générer la phrase cible (decoder).
L'architecture encoder-decoder est naturellement taillée pour ça.

## Exercise 3 : Bénéfices et Éthique du Pré-entraînement

### 5 bénéfices des modèles pré-entraînés

1. **Improved generalization** : Le modèle a vu tellement de textes
   variés qu'il généralise bien sur des données qu'il n'a jamais vues.
   Comme un étudiant qui a beaucoup lu — il s'adapte à de nouveaux sujets.

2. **Less labeled data** : Le pré-entraînement est non-supervisé
   (pas besoin de labels). Pour le fine-tuning, quelques centaines
   d'exemples labellisés suffisent là où un modèle classique en
   nécessiterait des milliers.

3. **Faster fine-tuning** : Le modèle a déjà appris la langue.
   Il ne reste qu'à l'orienter vers la tâche spécifique.
   Un fine-tuning peut prendre quelques heures au lieu de semaines.

4. **Transfer learning** : Les connaissances apprises sur une tâche
   (compréhension du langage) se transfèrent automatiquement à d'autres
   tâches (classification, traduction, résumé). Un seul modèle, mille usages.

5. **Robustness** : Entraîné sur des textes variés (formels, informels,
   techniques), le modèle gère mieux les fautes d'orthographe, les
   argots, et les formulations inhabituelles.

---

### Problèmes éthiques

- **Biais** : Si les données d'entraînement contiennent des stéréotypes
  (racisme, sexisme), le modèle les reproduit. Ex : associer certains
  métiers à un genre.

- **Désinformation** : Un LLM peut générer du contenu faux avec
  une grande confiance (hallucinations). Dangereux dans un contexte
  médical ou financier.

- **Vie privée** : Des données personnelles présentes dans les textes
  d'entraînement peuvent être mémorisées et restituées.

- **Mauvaise utilisation** : Génération de spam, de phishing,
  de deepfakes textuels à grande échelle.

---

### Stratégies de mitigation

- **Curation des données** : Filtrer les contenus toxiques,
  haineux ou biaisés avant l'entraînement.
- **RLHF** (Reinforcement Learning from Human Feedback) : Des humains
  notent les réponses pour orienter le modèle vers des comportements sûrs.
- **Audits réguliers** : Tester le modèle sur des benchmarks de biais
  et de toxicité avant déploiement.
- **Confidentialité différentielle** : Techniques mathématiques pour
  éviter que le modèle mémorise des données personnelles.
- **Filtres de sécurité** : Bloquer les requêtes et réponses dangereuses
  au niveau de l'API.

## Exercise 4 : Transformer Architecture Deep Dive

### Self-Attention et Multi-Head Attention

**Self-Attention** :
Pour chaque mot d'une phrase, on crée 3 vecteurs : Query (Q), Key (K),
Value (V). On calcule les scores Q·Kᵀ/√d_k pour mesurer l'importance
de chaque mot par rapport aux autres. On applique softmax pour obtenir
des poids entre 0 et 1, puis on multiplie par V pour obtenir la
représentation finale enrichie du contexte.

**Multi-Head Attention** :
Au lieu d'une seule attention, on en fait plusieurs en parallèle
(ex : 8 têtes). Chaque tête apprend à regarder des relations différentes :
une tête peut capturer les relations sujet-verbe, une autre les
co-références, une autre les relations sémantiques. On concatène
ensuite toutes les sorties. C'est plus riche qu'une seule attention.

**Exemple concret** :
Phrase : "Le médecin a dit à l'infirmière qu'elle était fatiguée."
- Tête 1 peut capturer que "elle" réfère à "l'infirmière" (coréférence)
- Tête 2 peut capturer la relation "médecin → dit" (sujet-verbe)
- Tête 3 peut capturer "fatiguée → infirmière" (adjectif-nom)

---

### Objectifs de pré-entraînement

**MLM (Masked Language Modeling)** : On masque 15% des mots et le
modèle doit les deviner en utilisant le contexte des deux côtés.
→ Résultat : compréhension bidirectionnelle. Utilisé par BERT.

**CLM (Causal Language Modeling)** : Le modèle prédit le prochain
mot en ne voyant que les mots précédents.
→ Résultat : génération de texte. Utilisé par GPT.

**Quand utiliser MLM** : Tâches de compréhension — classification,
NER, Q&A. Exemple : détecter si un email est du spam.

**Quand utiliser CLM** : Tâches de génération — chatbot, résumé,
traduction. Exemple : générer une réponse à un client.

**NSP (Next Sentence Prediction)** : BERT original prédisait si
deux phrases se suivaient réellement. Cela aidait pour les tâches
comme Q&A. Mais RoBERTa a montré que NSP n'améliorait pas vraiment
les performances — parfois il les dégradait. Les modèles modernes
l'ont abandonné pour se concentrer uniquement sur MLM.

---

### Sélection du type de modèle

| Tâche | Type choisi | Justification |
|-------|------------|---------------|
| Analyse sentiment avis clients | **Encoder-only** (BERT) | Tâche de compréhension : lire → classifier. Pas besoin de générer du texte. |
| Chatbot créatif | **Decoder-only** (GPT) | Génère du texte mot par mot de façon fluide et naturelle. |
| Traduction EN→ES | **Encoder-Decoder** (T5) | Comprendre la source (encoder) puis générer la cible (decoder). |

---

### Positional Encoding

**Pourquoi** : Le Transformer traite tous les mots en parallèle —
il ne sait pas naturellement quel mot vient avant ou après.
L'encodage positionnel ajoute une information sur la position
de chaque token dans la séquence.

**Sans positional encoding** : Les phrases "Le chat mange la souris"
et "La souris mange le chat" seraient identiques pour le modèle —
mêmes mots, même signification perçue. Ce qui est évidemment faux
et catastrophique pour la compréhension.

## Exercise 5 : BERT Variations

### Scénarios

- **Scénario 1** (mobile, ressources limitées) → **DistilBERT**
  40% plus léger que BERT, 60% plus rapide, conserve 97% des performances.
  Parfait pour tourner sur un téléphone avec peu de mémoire.

- **Scénario 2** (documents légaux, haute précision) → **RoBERTa**
  Entraîné plus longtemps sur plus de données, sans NSP qui introduisait
  du bruit. Donne les meilleures performances sur les tâches de compréhension.

- **Scénario 3** (support multilingue) → **XLM-RoBERTa**
  Entraîné sur 100 langues simultanément. Idéal pour du support client
  en français, anglais, dioula, etc.

- **Scénario 4** (pré-entraînement efficace + détection remplacement) → **ELECTRA**
  Utilise un générateur + discriminateur. Plus efficace que MLM :
  apprend sur TOUS les tokens, pas seulement les 15% masqués.

- **Scénario 5** (environnements contraints) → **ALBERT**
  Partage les paramètres entre les couches — beaucoup moins de mémoire
  pour des performances comparables à BERT.

---

### Tableau comparatif

| Modèle | Données/Méthode | Taille/Efficacité | Innovations | Cas d'usage idéaux |
|--------|----------------|-------------------|-------------|-------------------|
| RoBERTa | Plus de données, plus longtemps, sans NSP | Plus grand que BERT | Suppression NSP, dynamic masking | Tâches haute précision (legal, médical) |
| ALBERT | Mêmes données BERT | Très léger (partage de paramètres) | Parameter sharing, SOP au lieu NSP | Environnements mémoire limitée |

## Exercise 6 : Softmax Temperature

### 1) Scénarios de température

- **T = 0.2** : Le modèle devient très conservateur. Il choisit presque
  toujours le mot le plus probable. Le texte est cohérent et prévisible
  mais répétitif. Comme quelqu'un qui répond toujours "ça va" à "comment tu vas".

- **T = 1.5** : Le modèle devient très créatif — parfois trop. Il choisit
  des mots moins probables, ce qui donne un texte original et surprenant
  mais parfois incohérent ou hors sujet.

- **T = 1.0** : C'est la température par défaut — équilibre entre
  cohérence et variété. Le modèle suit la distribution originale
  apprise pendant l'entraînement.

---

### 2) Application Design

**Histoires pour enfants** :
J'utiliserais une température autour de **0.8 à 1.0**. Assez élevée
pour que les histoires soient créatives et variées (les enfants aiment
la surprise), mais pas trop pour que le texte reste cohérent et
compréhensible. Je pourrais augmenter légèrement (1.1) pour les
passages d'aventure et baisser (0.7) pour les conclusions morales.

**Résumés de rapports financiers** :
J'utiliserais une température très basse : **0.1 à 0.3**. Les chiffres
et faits financiers doivent être précis et fiables. Une température
élevée risquerait d'inventer des données ou de formuler des chiffres
incorrectement — ce qui serait catastrophique dans un contexte financier.

---

### 3) Température et Biais

Une température basse amplifie les biais présents dans les données
d'entraînement, car le modèle choisit toujours les mots les plus
probables — qui reflètent les stéréotypes les plus fréquents dans
les textes vus.

**Exemple concret** : Si on demande "Un bon ingénieur est..."
avec T=0.2, le modèle risque de compléter systématiquement avec
des pronoms masculins car c'est statistiquement plus fréquent
dans ses données d'entraînement. Avec T=1.2, il y aurait plus
de variété dans les réponses — réduisant légèrement ce biais
visible, même si le biais sous-jacent reste dans les poids du modèle.

### (Optional) Quick Generation Demo Across Temperatures
Note: Requires `transformers` and model download; skip if offline.


```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

model_name = "gpt2"
prompt     = "In Abidjan, the future of technology will"

pipe = pipeline(
    "text-generation",
    model=AutoModelForCausalLM.from_pretrained(model_name),
    tokenizer=AutoTokenizer.from_pretrained(model_name),
    device=0 if torch.cuda.is_available() else -1,
)

for T in [0.2, 1.0, 1.5]:
    out = pipe(
        prompt,
        max_new_tokens=40,
        temperature=T,
        do_sample=True,
        pad_token_id=pipe.tokenizer.eos_token_id
    )
    print(f"\n--- Temperature: {T} ---")
    print(out[0]["generated_text"])
```
