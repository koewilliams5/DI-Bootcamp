## Exercise 1: Rewrite and Optimize a Vague Prompt
**Scenario**: Manager asks for a LinkedIn post about FlowNest with the vague prompt `"Write something about productivity tips."`

### Tasks
- Identify 3 critical issues with the prompt.
- Rewrite it using: role prompting, clear instructions, and format/tone/length constraints.
- Add a final line to make it reusable for future social posts.

```python
# Three critical issues with the vague prompt
critical_issues = [
    'No brand voice or product mention: the prompt does not reference FlowNest, '
    'so the output could be generic advice unrelated to the app.',

    'No target audience: without specifying who the post is for (remote workers, '
    'students, managers…), the tone and examples will be unfocused.',

    'No format, length, or platform constraints: LinkedIn has a specific culture '
    '(professional yet engaging); without constraints the AI may produce a wall '
    'of text, bullet overload, or wrong character count.',
]
critical_issues
```

```python
# Rewritten prompt using role, instructions, and constraints
rewritten_prompt = """
You are a professional social media copywriter specializing in B2B SaaS startups.
Your task is to write a LinkedIn post promoting FlowNest, a focus and productivity
app designed for remote professionals and knowledge workers.

Instructions:
1. Open with a relatable pain point (e.g., constant context-switching or notification fatigue).
2. Introduce FlowNest as the solution in one sentence, highlighting its key benefit.
3. Give 3 concrete productivity tips that users can apply directly inside FlowNest.
4. Close with a clear call-to-action inviting readers to try FlowNest for free.

Constraints:
- Tone: professional yet warm and encouraging — avoid corporate jargon.
- Format: short intro paragraph + 3 bullet points + 1 CTA line.
- Length: 150–180 words total.
- Include 3 relevant hashtags at the end (e.g., #Productivity #RemoteWork #FlowNest).
- Do NOT use exclamation marks more than once.

Use this exact format every time we promote a productivity tool on LinkedIn,
replacing the app name, key benefit, and hashtags as needed.
"""
rewritten_prompt
```

**Why this works:**
- **Role prompting** sets expert context so the AI adopts the right voice.
- **Clear instructions** give a step-by-step structure that prevents random content.
- **Constraints** (word count, format, tone, hashtags) make the output immediately usable.
- **Reusability line** at the end turns this into a template for all future social posts.

## Exercise 2: Multi-Part Prompt for Quiz Generation
**Scenario**: Create an educational prompt for 7th graders about volcanic eruptions.

```python
# Full multi-part prompt
quiz_prompt = """
You are an enthusiastic middle-school science teacher creating engaging content
for a 7th-grade class (students aged 11–13).

I will provide you with a short article about volcanic eruptions.
Your task is to produce TWO sections of educational content:

--- SECTION 1: SUMMARY ---
Write exactly 2 bullet points summarising the most important ideas from the article.
- Use simple vocabulary appropriate for 11–13 year-olds (avoid technical jargon;
  if a scientific term is necessary, define it in parentheses).
- Each bullet must be a single sentence, maximum 20 words.

--- SECTION 2: MULTIPLE-CHOICE QUIZ ---
Generate exactly 3 multiple-choice questions (MCQs) based ONLY on information
found in the article. Do NOT invent facts.

For each question:
  - Write a clear, age-appropriate question.
  - Provide 3 answer options labelled A, B, C.
  - Mark the correct answer with ✅ and explain in one sentence why it is correct.
  - Make the 2 wrong options plausible but clearly distinguishable on reflection
    (these are distractors — avoid obviously silly answers).

--- OUTPUT FORMAT (Google Slides-ready) ---
Present the output using this exact structure so it can be pasted into Google Slides:

SLIDE 1 — KEY TAKEAWAYS
• [bullet 1]
• [bullet 2]

SLIDE 2 — QUESTION 1
[Question text]
A. ...
B. ...
C. ...
✅ Answer: [letter] — [one-sentence explanation]

(Repeat SLIDE 3 and SLIDE 4 for Questions 2 and 3)

Tone: friendly, encouraging, and fun — use short sentences and positive language.
"""
quiz_prompt
```

```python
# Three improvements over the vague prompt "Make a quiz for kids about this article."
improvements = [
    'Defined age group and vocabulary level: by specifying "11-13 year-olds" and '
    'requiring simple language with definitions for jargon, the prompt prevents the '
    'model from generating overly academic or overly infantile content.',

    'Structured multi-part output with exact counts: requesting "exactly 2 bullets" '
    'and "exactly 3 MCQs" with 1 correct answer and 2 specific distractors eliminates '
    'ambiguity — the model cannot skip sections or produce vague open-ended questions.',

    'Platform-specific format (Google Slides-ready): the slide-by-slide template with '
    'clear labels (SLIDE 1, SLIDE 2…) means the output can be pasted directly into '
    'a presentation without reformatting, saving time and avoiding layout errors.',
]
improvements
```

## Exercise 3: Add Context, Get Better Results
**Scenario**: Intern's vague request: `Summarize this report.`

```python
# Context table — what is missing and what should be added
context_table = {
    'Role': {
        'missing': True,
        'add': 'Act as a senior financial analyst with experience presenting '
               'data to executive leadership.'
    },
    'Audience': {
        'missing': True,
        'add': 'The summary will be presented to a non-technical management '
               'team during a 3-minute slot in the monthly all-hands meeting.'
    },
    'Purpose': {
        'missing': True,
        'add': 'The goal is to highlight the 3 most important financial '
               'findings and flag any risks that require immediate attention.'
    },
    'Input Source': {
        'missing': True,
        'add': 'The input is a 12-page Q2 financial report containing '
               'revenue, expenses, EBITDA, and cash-flow data.'
    },
    'Format/Style': {
        'missing': True,
        'add': 'Use 3 clearly labelled bullet points. Each point must cite '
               'a specific figure from the report. Use plain language — '
               'no financial jargon.'
    },
    'Constraints': {
        'missing': True,
        'add': 'Maximum 80 words total. Do not include information not '
               'present in the report. End with one sentence on recommended '
               'next steps.'
    },
}
context_table
```

```python
# Context-rich rewrite of the intern's prompt
finance_prompt = """
Act as a senior financial analyst who regularly briefs non-technical executives.

I will provide you with our company's Q2 financial report (12 pages, covering
revenue, expenses, EBITDA, and cash flow).

Your task: produce a concise executive summary for our monthly all-hands meeting.
The presenter will have exactly 3 minutes, so brevity is critical.

Format:
- Exactly 3 bullet points, each on a separate line.
- Each bullet must reference at least one specific figure or percentage from the report.
- Use plain, jargon-free language suitable for a non-finance audience.
- End with one sentence identifying the single most urgent risk and a recommended action.

Constraints:
- Maximum 80 words total (including the final risk sentence).
- Do NOT include any data or claims not explicitly present in the report.
- Do NOT use bullet sub-points or nested lists.

[PASTE REPORT TEXT HERE]
"""
finance_prompt
```

## Exercise 4: Match Prompt to Purpose
Pick a style (Exploratory, Structured, Conversational, Functional) to build a customer-support chatbot prompt.

```python
# Chosen style and full prompt
chosen_style = 'Structured'

prompt_template = """
You are a helpful customer support agent for ShopEasy, an e-commerce platform.
Your role is to resolve customer issues efficiently and politely.

When a customer contacts you, always follow this exact sequence:

STEP 1 — GREET AND ACKNOWLEDGE
Greet the customer by name (if available) and confirm you have understood their issue
in one sentence.

STEP 2 — CLASSIFY THE ISSUE
Internally identify which category applies (do NOT show this to the customer):
  [ORDER] — delayed, missing, or wrong item
  [RETURN] — refund or exchange request
  [ACCOUNT] — login, password, or billing issue
  [OTHER] — anything else

STEP 3 — PROVIDE SOLUTION
Give a clear, numbered resolution (maximum 3 steps). Each step must be actionable.
If you cannot resolve it, escalate with: "I'm connecting you to a specialist now."

STEP 4 — CONFIRM AND CLOSE
Ask: "Does this solve your issue, or is there anything else I can help with?"

Rules:
- Tone: calm, empathetic, and professional at all times.
- Never promise a refund without checking order status first.
- Response length: 60–100 words per reply.
"""

key_style_features = [
    'Fixed step sequence (STEP 1→4): the prompt enforces a rigid, repeatable '
    'workflow — greeting, classification, solution, close — which is the hallmark '
    'of Structured style.',

    'Numbered resolution steps and explicit labelling (e.g., [ORDER], [RETURN]): '
    'these predefined categories and ordered lists ensure every response follows '
    'the same predictable format, preventing free-form or rambling answers.',
]

justification = """
Structured style is the best fit for a customer-support chatbot because:
- Consistency is critical: customers expect the same quality of service every time,
  regardless of which interaction instance handles their query.
- Reduces agent errors: a fixed workflow prevents the AI from skipping essential steps
  (e.g., confirming identity before changing account details).
- Easier to audit and improve: structured outputs can be logged and compared against
  KPIs (resolution rate, word count compliance) in a way that conversational or
  exploratory outputs cannot.

Conversational style would feel natural but lacks the guardrails needed in a
commercial support context. Exploratory style is suited for brainstorming, not
problem resolution. Functional style executes tasks but doesn't handle the
empathy and multi-step dialogue that support requires.
"""

print(f"Chosen style: {chosen_style}")
print()
print("Key style features:")
for i, f in enumerate(key_style_features, 1):
    print(f"  {i}. {f}")
print()
print("Justification:", justification)
```

## Exercise 5: Prompt Refinement Challenge — Control Style, Structure, and Length
**Scenario**: Write a prompt for PulseOne Mini smartwatch email blurb.

```python
# Version 1 — initial prompt with strict constraints
pulseone_prompt_v1 = """
Write a product description for the PulseOne Mini smartwatch for a marketing email.

Constraints (follow all of them exactly):
- Tone: friendly and enthusiastic — write as if talking to a friend.
- Format: bullet points ONLY (no introductory paragraph, no closing sentence).
- Total length: strictly under 50 words — count every word before submitting.
- You MUST mention all three features in separate bullets:
    1. Battery life
    2. Fitness tracking
    3. Bluetooth connectivity
"""
pulseone_prompt_v1
```

```python
# Simulated model output for v1 (illustrative)
simulated_output_v1 = """
• Long-lasting battery — up to 7 days on a single charge, so you never miss a beat.
• Smart fitness tracking — monitors steps, heart rate, and sleep effortlessly.
• Instant Bluetooth sync — connects to your phone in seconds for calls and notifications.
"""

# Count words
word_count = len(simulated_output_v1.split())
features_covered = all(kw in simulated_output_v1.lower()
                       for kw in ['battery', 'fitness', 'bluetooth'])

# Self-evaluation
eval_v1 = {
    'word_count_ok': word_count <= 50,
    'actual_word_count': word_count,
    'features_covered': features_covered,
    'tone_ok': True,
    'notes': (
        f'Word count: {word_count} words — '
        + ('✅ within 50-word limit.' if word_count <= 50
           else '❌ exceeds limit — needs tightening.')
        + ' All 3 features mentioned: '
        + ('✅' if features_covered else '❌')
        + '. Tone is friendly and conversational: ✅'
    ),
}
print(eval_v1)
```

```python
# Version 2 — tighter constraints to guarantee compliance
pulseone_prompt_v2 = """
Write a product description for the PulseOne Mini smartwatch for a marketing email.

MANDATORY RULES — violating any rule means the output is invalid:
1. Write EXACTLY 3 bullet points — no more, no less.
2. Each bullet must cover EXACTLY ONE of these features (one bullet each):
   - Bullet 1: battery life
   - Bullet 2: fitness tracking
   - Bullet 3: Bluetooth connectivity
3. Each bullet must be 10–15 words long. Count the words in each bullet.
4. Total word count across all 3 bullets: do NOT exceed 45 words.
5. Tone: friendly and upbeat — use active verbs and avoid jargon.
6. Do NOT add a title, intro sentence, or closing remark.

Before submitting, verify: (a) bullet count = 3, (b) each feature appears once,
(c) total words ≤ 45.
"""
pulseone_prompt_v2
```

**Key refinements in v2:**
- Each bullet is assigned exactly one feature, preventing the model from grouping or omitting features.
- Per-bullet word count (10–15 words) adds a micro-constraint that forces conciseness at the sentence level, not just overall.
- The self-verification instruction (`Before submitting, verify…`) prompts the model to self-check before outputting.

## Exercise 6: Hallucination Spotting and Mitigation
**Scenario**: Model added an unsupported claim in a climate/marine article summary.

```python
# Baseline anti-hallucination prompt
no_hallucination_prompt = """
You are a scientific fact-checker tasked with summarising a peer-reviewed article.

I will provide you with the full text of the article below.
Your task is to write a summary of exactly 3 bullet points.

STRICT RULES:
- Use ONLY information that appears verbatim or is clearly implied in the article text.
- Do NOT add statistics, predictions, percentages, or conclusions
  that are not explicitly stated in the article.
- Do NOT rely on your prior training knowledge about climate change or marine biology.
  Treat the article as your sole source of truth.
- If the article does not provide enough information to fill 3 bullets,
  write fewer bullets rather than inventing content.

[PASTE ARTICLE TEXT HERE]
"""
no_hallucination_prompt
```

```python
# Verification-focused rewrite
verification_prompt = """
You are a rigorous scientific editor summarising a peer-reviewed article on
climate change and marine biodiversity.

Rules for every claim you include:
1. Only include claims EXPLICITLY stated in the article — not implied, inferred,
   or known from general knowledge.
2. After each bullet, add a parenthetical source tag: (Article, para. X) or
   (Article, Table Y) to show where the claim comes from.
3. If you are uncertain whether a claim is in the article, write:
   "[NOT FOUND IN ARTICLE]" instead of guessing.
4. Never use language like "studies suggest" or "it is expected" unless those
   exact words appear in the article.
5. After the summary, add a one-line disclaimer:
   "All claims above are drawn directly from the provided article text."

Format: exactly 3 bullet points (or fewer if the article lacks sufficient content).

[PASTE ARTICLE TEXT HERE]
"""
verification_prompt
```

```python
# Mitigation strategies and risky domains
mitigation_strategies = [
    'Source-anchoring constraint: explicitly instructing the model to use ONLY '
    'the provided text (and not its training knowledge) forces it to stay grounded. '
    'The instruction "do NOT rely on your prior training knowledge" directly '
    'counteracts the model's tendency to fill gaps with plausible but unverified data.',

    'Uncertainty signalling ("[NOT FOUND IN ARTICLE]"): requiring the model to flag '
    'missing information rather than fill it in transforms hallucination into a '
    'visible, auditable gap — making it easy for a human reviewer to spot and correct.',
]

risky_domains = {
    'domain_1': {
        'name': 'Healthcare / Medicine',
        'why': 'A hallucinated drug dosage, contraindication, or diagnosis can '
               'directly harm or kill a patient. Clinicians may not have time to '
               'verify every AI-generated claim, and a single confident-sounding '
               'error can have irreversible consequences.'
    },
    'domain_2': {
        'name': 'Legal / Compliance',
        'why': 'Fabricated case citations (e.g., "hallucinated" court rulings) '
               'have already led to lawyers being sanctioned for submitting '
               'non-existent precedents in court filings. Incorrect legal advice '
               'can lead to financial penalties, lost cases, or criminal liability.'
    },
}

print("Mitigation strategies:")
for i, s in enumerate(mitigation_strategies, 1):
    print(f"  {i}. {s}")
print()
print("Risky domains:")
for k, v in risky_domains.items():
    print(f"  [{v['name']}] {v['why']}")
```
