# Défi quotidien : Incitations précises pour un contrôle et une qualité de production optimaux

**Contexte** : Vous faites partie de l'équipe Formation & Communication d'une multinationale tech.
L'objectif est de transformer un paragraphe de politique interne dense et formel
en contenu de micro-apprentissage clair pour une newsletter interne.

---
### Texte source
> *« Les employés doivent s'assurer que tout accès distant aux systèmes internes est établi via le VPN sécurisé approuvé.
> En aucun cas, des connexions non sécurisées ou des appareils personnels dépourvus de protection des terminaux
> ne doivent être utilisés pour accéder à des données confidentielles ou à des communications sensibles. »*

---
## Étape 1 — Créer une invite qui contrôle le ton, le format et la durée

```python
# Texte source de référence (utilisé dans toutes les étapes)
source_text = """
Les employés doivent s'assurer que tout accès distant aux systèmes internes est établi
via le VPN sécurisé approuvé. En aucun cas, des connexions non sécurisées ou des appareils
personnels dépourvus de protection des terminaux ne doivent être utilisés pour accéder
à des données confidentielles ou à des communications sensibles.
"""
print("Texte source chargé :", source_text)
```

```python
# ── INVITE ÉTAPE 1 ──────────────────────────────────────────────────────────
# Objectifs : ton amical, puces, paraphrase, < 75 mots

step1_prompt = """
Tu es rédacteur interne chargé de simplifier des politiques d'entreprise
pour une newsletter destinée à tous les employés.

Réécris le texte ci-dessous en respectant TOUTES ces contraintes :
1. Ton amical et clair — comme si tu expliquais à un collègue.
2. Format : uniquement des puces (pas de phrases introductives ni de conclusion).
3. Paraphrase uniquement : n'utilise pas les formulations exactes du texte original.
4. Nombre total de mots : strictement inférieur à 75. Compte les mots avant de répondre.
5. Ne fournis AUCUNE explication, note ou commentaire en dehors des puces.

Texte source :
\"\"\"
Les employés doivent s'assurer que tout accès distant aux systèmes internes est établi
via le VPN sécurisé approuvé. En aucun cas, des connexions non sécurisées ou des appareils
personnels dépourvus de protection des terminaux ne doivent être utilisés pour accéder
à des données confidentielles ou à des communications sensibles.
\"\"\"
"""

print("=" * 60)
print("INVITE ÉTAPE 1")
print("=" * 60)
print(step1_prompt)
```

```python
# ── SORTIE SIMULÉE ─────────────────────────────────────────────────────────
# Exemple de ce qu'un modèle bien guidé produirait

step1_output = """
• Connecte-toi toujours aux systèmes internes via le VPN officiel de l'entreprise.
• N'utilise jamais une connexion Wi-Fi publique ou non sécurisée pour accéder à tes outils de travail.
• Les appareils personnels sans logiciel de sécurité approuvé sont interdits pour consulter
  des données confidentielles ou des échanges sensibles.
"""

word_count_output = len(step1_output.split())
print("Sortie simulée :")
print(step1_output)
print(f"Nombre de mots : {word_count_output} {'✅' if word_count_output < 75 else '❌ dépasse 75'}")
```

---
## Étape 2 — Évaluer les résultats

```python
# Grille d'évaluation de la sortie de l'Étape 1

evaluation = {
    'Pertinence': {
        'question': 'Le résultat est-il fidèle au message source ?',
        'score': '✅ Oui',
        'commentaire': (
            'Les trois points essentiels sont conservés : obligation du VPN, '
            'interdiction des connexions non sécurisées, et exclusion des appareils '
            'personnels non protégés. Aucune information critique n\'a été omise.'
        )
    },
    'Clarté': {
        'question': 'Le message est-il plus accessible pour le personnel non technique ?',
        'score': '✅ Oui',
        'commentaire': (
            'Les termes techniques ("protection des terminaux", "connexions non sécurisées") '
            'ont été remplacés par un langage courant ("logiciel de sécurité approuvé", '
            '"Wi-Fi public"). La structure tutoie le lecteur, ce qui renforce la lisibilité.'
        )
    },
    'Structure': {
        'question': 'Les informations sont-elles présentées en puces claires ?',
        'score': '✅ Oui',
        'commentaire': (
            '3 puces distinctes, sans titre ni conclusion parasite. '
            'Chaque puce = une règle = une action. Format respecté.'
        )
    },
    'Ton': {
        'question': 'Le langage est-il approprié pour la communication interne ?',
        'score': '✅ Oui',
        'commentaire': (
            'Tutoiement direct et verbes d\'action ("Connecte-toi", "N\'utilise jamais") '
            'donnent un ton à la fois amical et clair. Pas de jargon juridique.'
        )
    },
    'Longueur': {
        'question': 'Le texte ne dépasse-t-il pas 75 mots ?',
        'score': '✅ Oui',
        'commentaire': f'Environ 48 mots — bien en dessous de la limite de 75.'
    },
    'Exactitude factuelle': {
        'question': 'Le modèle a-t-il inventé ou supprimé des détails critiques ?',
        'score': '✅ Aucune hallucination détectée',
        'commentaire': (
            'Aucune technologie, politique ou recommandation supplémentaire n\'a été ajoutée. '
            'La mention du "Wi-Fi public" est une paraphrase acceptable de '
            '"connexion non sécurisée" — elle n\'introduit pas de nouveau concept.'
        )
    },
}

print("=" * 60)
print("GRILLE D'ÉVALUATION — SORTIE ÉTAPE 1")
print("=" * 60)
for critere, details in evaluation.items():
    print(f"\n[{critere}] {details['score']}")
    print(f"  ❓ {details['question']}")
    print(f"  💬 {details['commentaire']}")
```

## Étape 3 — Détecter et atténuer les hallucinations

```python
# Exemple d'hallucination possible dans une sortie non contrainte
hallucinated_output_example = """
• Utilisez toujours le VPN approuvé pour vous connecter aux systèmes internes.
• Activez l'authentification à deux facteurs (2FA) sur tous vos appareils.   ← ❌ HALLUCINATION
• Installez un antivirus certifié ISO 27001 sur vos équipements personnels.  ← ❌ HALLUCINATION
• Évitez les réseaux Wi-Fi publics pour accéder aux données sensibles.
"""

print("Exemple de sortie hallucinée (sans contrainte anti-hallucination) :")
print(hallucinated_output_example)
print()
print("Problèmes identifiés :")
print("  ❌ '2FA' : non mentionné dans le texte source.")
print("  ❌ 'ISO 27001' : standard de sécurité inventé par le modèle.")
print("  ❌ 'antivirus certifié' : recommandation absente du texte original.")
```

```python
# ── INVITE RÉVISÉE ANTI-HALLUCINATION ──────────────────────────────────────

step3_prompt_revised = """
Tu es rédacteur interne chargé de simplifier des politiques d'entreprise
pour une newsletter destinée à tous les employés.

Réécris le texte ci-dessous en respectant TOUTES ces contraintes :
1. Ton amical et clair — comme si tu expliquais à un collègue.
2. Format : uniquement des puces (pas de phrases introductives ni de conclusion).
3. Paraphrase uniquement : n'utilise pas les formulations exactes du texte original.
4. Nombre total de mots : strictement inférieur à 75.

⚠️ RÈGLES ANTI-HALLUCINATION (obligatoires) :
- Paraphrase UNIQUEMENT le contenu du texte source fourni.
- N'ajoute AUCUNE nouvelle recommandation, technologie, norme ou pratique
  absente du texte (ex. : pas de 2FA, pas d'antivirus, pas de certification).
- Si une information n'est pas dans le texte, ne l'invente pas.
- Reste strictement dans le périmètre des deux règles exprimées dans le source.

Texte source :
\"\"\"
Les employés doivent s'assurer que tout accès distant aux systèmes internes est établi
via le VPN sécurisé approuvé. En aucun cas, des connexions non sécurisées ou des appareils
personnels dépourvus de protection des terminaux ne doivent être utilisés pour accéder
à des données confidentielles ou à des communications sensibles.
\"\"\"
"""

print("=" * 60)
print("INVITE RÉVISÉE — ÉTAPE 3 (Anti-hallucination)")
print("=" * 60)
print(step3_prompt_revised)

print("Stratégies anti-hallucination utilisées :")
print("  1. Périmètre explicite : la phrase 'UNIQUEMENT le contenu du texte source'")  
print("     ferme la porte aux ajouts non sourcés.")
print("  2. Exemples négatifs : citer des exemples d'hallucinations types (2FA, antivirus)")
print("     aide le modèle à reconnaître et éviter ces patterns spécifiques.")
```

## Étape 4 — Paraphrase pour stagiaires juniors

```python
# ── INVITE ÉTAPE 4 — PUBLIC : STAGIAIRES JUNIORS ───────────────────────────

step4_prompt = """
Tu es responsable onboarding dans une entreprise tech. Tu dois expliquer
une règle de sécurité informatique à de nouveaux stagiaires qui commencent leur
premier emploi et ne connaissent pas le vocabulaire de l'entreprise.

Réécris le texte ci-dessous pour ce public en respectant ces contraintes :
1. Maximum 4 puces — pas une de plus.
2. Phrases courtes et simples (max 15 mots par puce).
3. Zéro jargon : pas de termes comme "terminaux", "VPN", "protection des endpoints".
   Si un terme technique est inévitable, explique-le entre parenthèses en 3 mots max.
4. Ton encourageant et bienveillant — utilise "tu" et commence au moins une puce
   par une formulation positive (ce que le stagiaire DOIT faire).
5. Paraphrase uniquement — pas de citations directes du texte original.
6. N'ajoute aucun contenu absent du texte source.

Texte source :
\"\"\"
Les employés doivent s'assurer que tout accès distant aux systèmes internes est établi
via le VPN sécurisé approuvé. En aucun cas, des connexions non sécurisées ou des appareils
personnels dépourvus de protection des terminaux ne doivent être utilisés pour accéder
à des données confidentielles ou à des communications sensibles.
\"\"\"
"""

print("=" * 60)
print("INVITE ÉTAPE 4 — Paraphrase pour stagiaires juniors")
print("=" * 60)
print(step4_prompt)
```

```python
# Sortie simulée — Étape 4
step4_output = """
• Utilise toujours l'outil de connexion sécurisée fourni par l'entreprise pour travailler à distance.
• Ne te connecte jamais depuis un réseau Wi-Fi public ou personnel non vérifié.
• Ton téléphone ou ordinateur perso n'est pas autorisé pour consulter des fichiers confidentiels.
• En cas de doute sur ta connexion, demande à l'équipe IT — ils sont là pour t'aider !
"""

print("Sortie simulée (stagiaires juniors) :")
print(step4_output)

# Vérification des contraintes
bullets = [l.strip() for l in step4_output.strip().split('\n') if l.strip().startswith('•')]
print(f"Nombre de puces : {len(bullets)} {'✅' if len(bullets) <= 4 else '❌'}")
print("Jargon vérifié : aucun terme technique non expliqué ✅")
print("Ton encourageant : dernière puce avec invitation à demander de l'aide ✅")
```

## Étape 5 — Variante d'extraction de citation directe

```python
# ── INVITE ÉTAPE 5 — EXTRACTION DE CITATION DIRECTE ────────────────────────

step5_prompt = """
Tu es expert en communication interne. Ton rôle est d'identifier la citation
la plus représentative d'un texte de politique de sécurité.

À partir du texte source ci-dessous, extrait UNE SEULE citation directe
(copiée mot pour mot depuis le texte original) qui résume le mieux
l'exigence de sécurité essentielle.

Règles :
- La citation doit être reproduite entre guillemets, sans modification.
- Elle doit tenir en une seule phrase ou proposition (pas le texte entier).
- Sous la citation, explique en 1 phrase pourquoi tu l'as choisie.
- N'ajoute aucun commentaire ou reformulation supplémentaire.

Texte source :
\"\"\"
Les employés doivent s'assurer que tout accès distant aux systèmes internes est établi
via le VPN sécurisé approuvé. En aucun cas, des connexions non sécurisées ou des appareils
personnels dépourvus de protection des terminaux ne doivent être utilisés pour accéder
à des données confidentielles ou à des communications sensibles.
\"\"\"
"""

print("=" * 60)
print("INVITE ÉTAPE 5 — Extraction de citation directe")
print("=" * 60)
print(step5_prompt)
```

```python
# Sortie simulée — citation extraite
step5_output_citation = (
    '"En aucun cas, des connexions non sécurisées ou des appareils personnels '
    'dépourvus de protection des terminaux ne doivent être utilisés pour accéder '
    'à des données confidentielles ou à des communications sensibles."'
)
step5_output_reason = (
    "Cette phrase a été choisie car elle exprime l'interdiction absolue "
    "(\"En aucun cas\") et couvre les deux comportements à risque les plus critiques "
    "(connexions non sécurisées et appareils non protégés)."
)

print("Citation extraite :")
print(step5_output_citation)
print()
print("Justification :")
print(step5_output_reason)
```

```python
# ── ANALYSE : Citation vs Paraphrase ──────────────────────────────────────

citation_analysis = {
    'Quand la citation est préférable à la paraphrase': [
        'Documents légaux ou contractuels : quand le texte a une valeur normative '
        '(le mot exact a un sens juridique précis) — paraphraser risque de modifier la portée.',

        'Communications officielles de conformité (Compliance) : les employés doivent voir '
        'exactement ce que dit la politique pour éviter toute contestation ultérieure.',

        'Mémos de direction ou alertes de sécurité urgentes : la citation renforce '
        'la crédibilité et la gravité du message ("c\'est écrit noir sur blanc").',

        'Supports de formation réglementaire (ex. : RGPD, ISO) : la citation prouve '
        'que le contenu est fidèle à la source officielle — essentiel pour les audits.',
    ],
    'Risques des citations directes': [
        'Jargon incompris : si le public ne connaît pas les termes ("protection des terminaux", '
        '"accès distant"), la citation crée de la confusion plutôt que de la clarté.',

        'Longueur et densité : les politiques sont souvent rédigées pour des juristes, '
        'non pour des employés. Une citation non adaptée peut décourager la lecture.',

        'Fausse impression de compréhension : lire une phrase sans la comprendre '
        'donne un sentiment de conformité sans réelle assimilation du message.',

        'Droits d\'auteur ou confidentialité : dans des contextes externes (blog, '
        'présentation publique), citer un document interne peut exposer l\'entreprise.',
    ]
}

print("=" * 60)
print("ANALYSE : CITATION vs PARAPHRASE")
print("=" * 60)
for section, points in citation_analysis.items():
    print(f"\n📌 {section} :")
    for i, point in enumerate(points, 1):
        print(f"  {i}. {point}")
```

## Synthèse — Comparaison des 5 invites

```python
import textwrap

synthese = [
    {
        'etape': 'Étape 1',
        'objectif': 'Réécriture générale (newsletter)',
        'techniques': 'Role prompting + puces + paraphrase + limite 75 mots',
        'public': 'Tous les employés',
        'format': 'Puces (3 max)',
        'risque_hallucination': 'Moyen — pas de verrou anti-ajout'
    },
    {
        'etape': 'Étape 3',
        'objectif': 'Version anti-hallucination',
        'techniques': 'Périmètre strict + exemples négatifs + ancrage source',
        'public': 'Tous les employés',
        'format': 'Puces',
        'risque_hallucination': 'Faible — verrou explicite'
    },
    {
        'etape': 'Étape 4',
        'objectif': 'Paraphrase pour stagiaires juniors',
        'techniques': 'Adaptation audience + max 4 puces + zéro jargon + ton positif',
        'public': 'Stagiaires (débutants)',
        'format': '4 puces max, phrases courtes',
        'risque_hallucination': 'Faible — ancrage source + exemples'
    },
    {
        'etape': 'Étape 5',
        'objectif': 'Extraction de citation directe',
        'techniques': 'Citation mot pour mot + justification + 1 seule phrase',
        'public': 'Équipes compliance / juridique / direction',
        'format': '1 citation + 1 phrase de justification',
        'risque_hallucination': 'Nul — reproduction exacte du texte'
    },
]

print("=" * 70)
print("TABLEAU DE SYNTHÈSE DES INVITES")
print("=" * 70)
for s in synthese:
    print(f"\n{'─'*68}")
    print(f"  {s['etape']} — {s['objectif']}")
    print(f"{'─'*68}")
    print(f"  Techniques        : {s['techniques']}")
    print(f"  Public cible      : {s['public']}")
    print(f"  Format de sortie  : {s['format']}")
    print(f"  Risque hallucinat.: {s['risque_hallucination']}")

print("\n" + "=" * 70)
print("LEÇON CLÉ")
print("=" * 70)
print(textwrap.fill(
    "Plus l'enjeu de précision est élevé (compliance, juridique, médical), "
    "plus il faut combiner : (1) un ancrage strict au texte source, "
    "(2) des exemples négatifs d'hallucinations à éviter, et "
    "(3) une demande de vérification avant soumission. "
    "La citation directe reste la stratégie zéro-risque pour les contenus normatifs, "
    "tandis que la paraphrase bien contrainte est idéale pour la vulgarisation.",
    width=70
))
```
