Défi quotidien : Intégration de NumPy, Pandas et Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Préparation des données
temperature = np.random.uniform(low= -5, high=35, size=(10, 12))

ville = ["Yamoussoukro", "Abidjan", "San-Pedro", "Bouaké", "Daloa", "Bouaflé", "Aboisso", "Dabou", "Korhogo", "Man"]
mois = ["Jan","Fev", "Mar", "Avr", "Mai", "Juin", "Juiel", "Août", "Septembre", "Octobre", "Novembre", "Decembre"]

df = pd.DataFrame(data= temperature, index= ville, columns=mois )

display(df)

#Analyse des données

# Moyenne par mois (axis=0 = on calcule sur les 10 villes de chaque colonne)
temperature_moyenne = df.mean(axis=0)

# Moyenne par ville (axis=1 = on calcule sur les 12 mois de chaque ligne)
temperature_moyenne_anuel = df.mean(axis=1)

# Ville la plus chaude et la plus froide
moyenne_max = temperature_moyenne_anuel.idxmax()
moyenne_min = temperature_moyenne_anuel.idxmin()

print(f"Ville la plus chaude : {moyenne_max} ({temperature_moyenne_anuel[moyenne_max]:.1f} C)")
print(f"Ville la plus froide : {moyenne_min} ({temperature_moyenne_anuel[moyenne_min]:.1f} C)")

#Visualisation des données

plt.Figure(figsize= (14, 7))
# len(villes) fonctionne car "villes" est toujours la liste originale
palette_verte = plt.cm.winter(np.linspace(0, 1, len(ville)))

# "nom_ville" evite d'ecraser la variable "villes"
for i, (nom_ville, temperatures) in enumerate(df.iterrows()):
    style = '--' if i % 2 == 0 else '-'
    plt.plot(mois, temperatures,
             marker='o',
             linewidth=1.5,
             markersize=4,
             linestyle=style,
             color=palette_verte[i],
             label=nom_ville)

plt.title("Monthly Temperature Trends for Each City",
          fontsize=16, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
