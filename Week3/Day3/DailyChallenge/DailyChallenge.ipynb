```python

# ======================================================
# 1. IMPORTATION DES LIBRAIRIES
# ======================================================

# Pandas : manipulation et analyse des données
import pandas as pd

# NumPy : calculs mathématiques et statistiques
import numpy as np

# SciPy : statistiques avancées
from scipy import stats

# Matplotlib : visualisation des données
import matplotlib.pyplot as plt

# Permet d'afficher les graphiques dans le notebook
%matplotlib inline

print("Libraries imported successfully")


# ======================================================
# 2. DATA LOADING AND EXPLORATION
# ======================================================

# Chargement du dataset
df = pd.read_csv(r"E:\FORMATION_Tech_Talent_Accelerator_de_initiative_FORPRODE\DI_Bootcamp\Week3\Day3\DailyChallenge\Mobile Price Classification\train.csv")

# Affichage des 5 premières lignes
print("First 5 rows of the dataset:")
display(df.head())

# Dimensions du dataset
print("\nDataset Shape:")
print(df.shape)

# Informations générales
print("\nDataset Information:")
df.info()

# Types des colonnes
print("\nData Types:")
print(df.dtypes)

# Description statistique
print("\nDescriptive Statistics:")
display(df.describe())


# ======================================================
# 3. DATA CLEANING AND PREPROCESSING
# ======================================================

# Vérification des valeurs manquantes
print("\nMissing Values:")
print(df.isnull().sum())

# Si des valeurs manquantes existent,
# elles sont remplacées par la moyenne
df.fillna(df.mean(numeric_only=True), inplace=True)

# Vérification des doublons
duplicates = df.duplicated().sum()

print("\nNumber of duplicate rows:", duplicates)

# Suppression des doublons
df.drop_duplicates(inplace=True)

print("Duplicates removed successfully.")

# Vérification finale
print("\nDataset Shape After Cleaning:")
print(df.shape)


# ======================================================
# 4. TARGET VARIABLE ANALYSIS
# ======================================================

print("\nDistribution of Price Range:")
print(df["price_range"].value_counts())

plt.figure(figsize=(8,5))

df["price_range"].value_counts().plot(kind="bar")

plt.title("Distribution of Mobile Price Categories")
plt.xlabel("Price Range")
plt.ylabel("Number of Phones")

plt.show()

# Commentaire :
# Cette visualisation permet de vérifier si les classes
# sont équilibrées ou non.


# ======================================================
# 5. DETAILED STATISTICAL ANALYSIS
# ======================================================

print("\nDETAILED STATISTICAL ANALYSIS")

for col in df.columns:

    if col != "price_range":

        print("\n")
        print("="*60)
        print("FEATURE:", col)
        print("="*60)

        # Mesures de tendance centrale
        mean = np.mean(df[col])
        median = np.median(df[col])
        mode = stats.mode(df[col], keepdims=True)

        print("Mean:", round(mean,2))
        print("Median:", round(median,2))
        print("Mode:", mode.mode[0])

        # Mesures de dispersion
        minimum = np.min(df[col])
        maximum = np.max(df[col])

        print("Minimum:", minimum)
        print("Maximum:", maximum)

        data_range = maximum - minimum

        print("Range:", data_range)

        variance = np.var(df[col])
        std_dev = np.std(df[col])

        print("Variance:", round(variance,2))
        print("Standard Deviation:", round(std_dev,2))

        # Forme de distribution
        skewness = stats.skew(df[col])
        kurtosis = stats.kurtosis(df[col])

        print("Skewness:", round(skewness,2))
        print("Kurtosis:", round(kurtosis,2))

# Commentaire :
# Mean      -> moyenne
# Median    -> valeur centrale
# Mode      -> valeur la plus fréquente
# Variance  -> dispersion des données
# Std Dev   -> écart-type
# Skewness  -> asymétrie
# Kurtosis  -> concentration autour de la moyenne


# ======================================================
# 6. FEATURE-TARGET CORRELATION ANALYSIS
# ======================================================

correlation_matrix = df.corr()

price_correlations = (
    correlation_matrix["price_range"]
    .sort_values(ascending=False)
)

print("\nCorrelation with Price Range:")
print(price_correlations)

# Commentaire :
# Les variables ayant les corrélations les plus fortes
# sont généralement les plus importantes pour prédire
# le prix d'un téléphone.


# ======================================================
# 7. HYPOTHESIS TESTING (ANOVA)
# ======================================================

# Exemple avec la RAM

group0 = df[df["price_range"] == 0]["ram"]
group1 = df[df["price_range"] == 1]["ram"]
group2 = df[df["price_range"] == 2]["ram"]
group3 = df[df["price_range"] == 3]["ram"]

f_statistic, p_value = stats.f_oneway(
    group0,
    group1,
    group2,
    group3
)

print("\nANOVA TEST ON RAM")

print("F-statistic:", f_statistic)
print("P-value:", p_value)

if p_value < 0.05:
    print("Result: RAM significantly influences mobile price.")
else:
    print("Result: RAM does not significantly influence mobile price.")

# Commentaire :
# Si p-value < 0.05
# alors la différence entre les groupes est significative.


# ======================================================
# 8. HISTOGRAM - BATTERY POWER
# ======================================================

plt.figure(figsize=(8,5))

plt.hist(
    df["battery_power"],
    bins=20
)

plt.title("Battery Power Distribution")
plt.xlabel("Battery Power")
plt.ylabel("Frequency")

plt.show()

# Commentaire :
# Permet de visualiser la répartition des valeurs.


# ======================================================
# 9. SCATTER PLOT - RAM VS PRICE RANGE
# ======================================================

plt.figure(figsize=(8,5))

plt.scatter(
    df["ram"],
    df["price_range"]
)

plt.title("RAM vs Price Range")
plt.xlabel("RAM")
plt.ylabel("Price Range")

plt.show()

# Commentaire :
# Permet d'observer la relation entre RAM
# et catégorie de prix.


# ======================================================
# 10. BOXPLOT - RAM
# ======================================================

plt.figure(figsize=(8,5))

plt.boxplot(df["ram"])

plt.title("RAM Boxplot")

plt.show()

# Commentaire :
# Permet d'identifier les valeurs aberrantes
# (outliers).


# ======================================================
# 11. HEATMAP OF CORRELATIONS
# ======================================================

plt.figure(figsize=(14,10))

heatmap = plt.imshow(
    correlation_matrix
)

plt.colorbar(heatmap)

plt.xticks(
    range(len(correlation_matrix.columns)),
    correlation_matrix.columns,
    rotation=90
)

plt.yticks(
    range(len(correlation_matrix.columns)),
    correlation_matrix.columns
)

plt.title("Correlation Heatmap")

plt.show()

# Commentaire :
# Plus la couleur est intense,
# plus la corrélation est forte.


# ======================================================
# 12. FEATURE IMPORTANCE
# ======================================================

importance = abs(
    df.corr()["price_range"]
)

importance = (
    importance.sort_values(
        ascending=False
    )
)

print("\nFeature Importance:")
print(importance)

plt.figure(figsize=(12,6))

importance.plot(
    kind="bar"
)

plt.title("Feature Importance Based on Correlation")
plt.xlabel("Features")
plt.ylabel("Absolute Correlation")

plt.show()

# Commentaire :
# Les variables ayant les barres les plus élevées
# sont les plus influentes sur le prix.


# ======================================================
# 13. CONCLUSION
# ======================================================

print("\nCONCLUSION")

print("""
1. The dataset was successfully loaded and cleaned.

2. No major missing values were found.

3. RAM is one of the strongest predictors
   of mobile price.

4. Battery Power also contributes to
   price classification.

5. Correlation analysis confirms that
   several hardware specifications
   influence mobile prices.

6. ANOVA testing demonstrated that
   differences between price groups
   are statistically significant.

7. Visualizations helped identify
   relationships, distributions,
   and important predictive features.
""")
```
