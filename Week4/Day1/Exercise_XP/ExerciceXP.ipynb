Exercice 1 : Définition du problème et collecte de données pour la prédiction des défauts de paiement des prêts

Problème : Une banque comme Ecobank CI reçoit des centaines de demandes de prêt chaque mois. Elle veut automatiquement prédire si un demandeur risque de ne pas rembourser son prêt (défaut de paiement), afin de réduire ses pertes financières. C'est un problème de classification binaire : la sortie est soit "défaut probable" soit "remboursement probable".


Données à collecter : Revenus mensuels, type d'emploi, montant du prêt, durée souhaitée, historique de crédit, dettes actuelles, nombre de personnes à charge, localité, dépôts bancaires mensuels.


Sources : Dossiers internes de la banque, bureau de crédit (comme la Centrale des Risques de la BCEAO en Côte d'Ivoire), formulaires de demande de prêt.

Exercise 2 : Feature Selection and Model Choice for Loan Default Prediction

Dataset utilisé : Loan Prediction Dataset (614 entrées, 13 colonnes)

Features sélectionnées et justifications
1. Credit_History
C'est la feature la plus importante du dataset. Elle indique si le client a déjà honoré ses dettes par le passé (1 = bon historique, 0 = mauvais). Une banque comme Ecobank se base énormément sur cet historique pour accorder ou refuser un prêt. Un client qui n'a jamais remboursé ses dettes passées a beaucoup plus de chances de faire défaut.
2. ApplicantIncome
Le revenu du demandeur est directement lié à sa capacité de remboursement. Plus le revenu est élevé, plus la personne peut rembourser ses mensualités sans difficulté.
3. CoapplicantIncome
Le revenu du co-demandeur (souvent le conjoint) s'ajoute au revenu principal. Un revenu combiné élevé rassure la banque sur la solidité financière du foyer.
4. LoanAmount
Le montant du prêt demandé influence directement le risque. Un prêt très élevé par rapport aux revenus du demandeur est plus difficile à rembourser.
5. Loan_Amount_Term
La durée de remboursement change le montant des mensualités. Un prêt sur 360 mois (30 ans) sera moins contraignant chaque mois qu'un prêt sur 12 mois.
6. Married
Une personne mariée dispose souvent de deux sources de revenus dans le foyer, ce qui augmente la capacité globale de remboursement.
7. Dependents
Plus une personne a de personnes à charge (enfants, parents), plus ses dépenses mensuelles sont élevées et moins elle dispose de ressources pour rembourser.
8. Education
Un diplômé accède généralement à des emplois plus stables et mieux rémunérés, ce qui réduit le risque de défaut de paiement.
9. Self_Employed
Un travailleur indépendant a un revenu irrégulier contrairement à un salarié fixe. Cette instabilité représente un risque supplémentaire pour la banque.
10. Property_Area
La zone géographique (urbaine, semi-urbaine, rurale) reflète le niveau d'accès à l'emploi et le coût de la vie. Une personne en zone urbaine a généralement plus d'opportunités économiques.


Features supprimées et justifications
    Loan_ID : C'est un simple identifiant de dossier. Il n'apporte aucune information sur le comportement financier du client et ne peut donc rien prédire.
    Gender
    Le genre d'une personne n'est pas un facteur déterminant de sa capacité à rembourser un prêt. L'utiliser dans un modèle ML pourrait introduire des biais discriminatoires. De plus, cette colonne contient 13 valeurs manquantes dans le dataset.


Observation sur les données manquantes
    Le dataset contient plusieurs valeurs manquantes notamment : Credit_History (50 manquantes), Self_Employed (32), LoanAmount (22), Dependents (15), Loan_Amount_Term (14). Ces valeurs devront être traitées avant l'entraînement du modèle, par exemple en les remplaçant par la moyenne ou la valeur la plus fréquente selon la colonne concernée.

Variable cible
    Loan_Status est la variable que le modèle doit prédire (Y = approuvé, N = refusé). Ce n'est pas une feature d'entrée. Le dataset contient 422 cas approuvés contre 192 refusés, soit un déséquilibre à prendre en compte lors de l'évaluation du modèle.





Exercice 3 : Entraînement, évaluation et optimisation du modèle

- Modèles choisis : Régression Logistique, Random Forest, et KNN.
La Régression Logistique est adaptée car c'est un problème de classification binaire. Random Forest est plus robuste grâce au vote de plusieurs arbres. KNN peut servir de comparaison.


- Évaluation : On utilise l'Accuracy pour une vue globale, mais surtout le Recall pour minimiser les cas où on accorde un prêt à quelqu'un qui ne remboursera pas. Le F1-Score équilibre les deux. On peut aussi utiliser la Cross-Validation pour tester le modèle sur plusieurs sous-ensembles des données et éviter le surapprentissage.

Exercice 4 : Concevoir des solutions d'apprentissage automatique pour des problèmes spécifiques

Scénario 1 — Prédire le prix des actions
Type : Apprentissage supervisé, spécifiquement de la régression.
Justification : On dispose de données historiques avec des prix connus (les labels). Le modèle apprend la relation entre les données passées et les prix pour prédire une valeur numérique continue dans le futur. Les prix étant des nombres et non des catégories, c'est une régression et non une classification.

Scénario 2 — Organiser une bibliothèque
Type : Apprentissage non supervisé, spécifiquement du clustering.
Justification : Aucun genre n'est défini à l'avance, donc il n'y a pas de labels. Le modèle analyse les similarités entre les livres (contenu, mots-clés, style) et les regroupe automatiquement en clusters. L'algorithme découvre lui-même la structure cachée dans les données.

Scénario 3 — Robot dans un labyrinthe
Type : Apprentissage par renforcement (Reinforcement Learning).
Justification : Le robot apprend par essais et erreurs dans un environnement inconnu. Chaque bonne action est récompensée, chaque mauvaise est pénalisée. Au fil du temps il optimise sa stratégie pour maximiser les récompenses et trouver le chemin le plus court. Il n'y a pas de données d'entraînement préexistantes — l'agent apprend en interagissant avec l'environnement.

Exercice 5 : Concevoir une stratégie d'évaluation pour différents modèles d'apprentissage automatique

1. Modèle Supervisé — Classification (Loan Default Prediction)
Pour évaluer ce modèle on utilise plusieurs métriques complémentaires.
L'Accuracy mesure le pourcentage de prédictions correctes sur l'ensemble des cas. Cependant elle ne suffit pas seule car notre dataset est déséquilibré (422 approuvés contre 192 refusés). Un modèle qui dirait toujours "approuvé" obtiendrait automatiquement 69% d'accuracy sans rien apprendre.
La Précision mesure si le modèle a raison quand il prédit qu'un client ne remboursera pas. Elle évite de rejeter de bons clients.
Le Recall mesure si le modèle détecte tous les mauvais payeurs. C'est la métrique la plus importante pour une banque car accorder un prêt à quelqu'un qui ne remboursera pas représente une perte financière directe.
Le F1-Score équilibre la Précision et le Recall. Il est utile quand les deux types d'erreurs sont importants.
On utilise également la Cross-Validation pour tester le modèle sur plusieurs sous-ensembles des données et s'assurer qu'il ne mémorise pas les données d'entraînement (surapprentissage).
Limite principale : sur un dataset déséquilibré, l'accuracy est trompeuse. Il faut toujours privilégier le Recall et le F1-Score pour ce type de problème.
2. Modèle Non Supervisé — Clustering
Sans labels prédéfinis, on ne peut pas comparer les résultats à une "bonne réponse". On utilise donc des métriques internes.
Le Silhouette Score mesure la qualité du placement de chaque point dans son groupe. Un point bien classé doit être proche des membres de son groupe et loin des autres groupes. Le score varie entre -1 et 1. Plus il est proche de 1, meilleur est le clustering.
L'Elbow Method aide à choisir le bon nombre de groupes. On teste plusieurs valeurs (2, 3, 4, 5 groupes...) et on observe le point où ajouter un groupe supplémentaire n'améliore plus significativement les résultats. Sur un graphique ce point forme un coude.
Limite principale : il n'existe pas de vérité absolue en clustering. Deux algorithmes différents peuvent produire des groupes différents, tous deux valides selon l'objectif.
3. Modèle de Reinforcement Learning
On évalue un agent de Reinforcement Learning avec trois indicateurs.
La Cumulative Reward est la somme totale des points gagnés sur une session. Si cette somme augmente régulièrement au fil des sessions, l'agent apprend et s'améliore.
La Convergence indique que l'agent a trouvé une stratégie stable et optimale. Il n'explore plus au hasard mais exploite ce qu'il a appris. On observe la convergence quand les performances se stabilisent.
L'équilibre Exploration vs Exploitation est crucial. Au début l'agent explore l'environnement au hasard pour découvrir les possibilités. Avec le temps il exploite les meilleures stratégies découvertes. Un bon modèle équilibre les deux — trop d'exploration ralentit l'apprentissage, trop d'exploitation empêche de découvrir de meilleures solutions.
Limite principale : l'entraînement peut être très long et coûteux en ressources. De plus l'agent peut converger vers une solution locale qui n'est pas forcément la meilleure solution globale.
